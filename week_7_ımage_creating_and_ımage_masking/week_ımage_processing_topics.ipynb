```python
import cv2
import matplotlib.pyplot as plt
import numpy as np;
```

```python
width,height,channel=300,300,3

img_red=np.zeros((width,height,channel),dtype=np.uint8)
img_red[:,:,0]=255
plt.imshow(img_red)
plt.axis('off')
```

```python
width,height,channel=300,300,3

img_green=np.zeros((width,height,channel),dtype=np.uint8)
img_green[:,:,1]=255
plt.imshow(img_green)
plt.axis('off')
```

```python
width,height,channel=300,300,3

img_blue=np.zeros((width,height,channel),dtype=np.uint8)
img_blue[:,:,2]=255
plt.imshow(img_blue)
plt.axis('off')
```

```python
fig=plt.figure(figsize=(10,7));
rows=1
columns=3

fig.add_subplot(rows,columns,1)
plt.imshow(img_red)
plt.axis('off')
plt.title("RED IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(img_green)
plt.axis('off')
plt.title("Green IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(img_blue)
plt.axis('off')
plt.title("BLUE IMAGE")
```

```python
img_black=np.zeros((width,height,3),dtype=np.uint8)
img_white=np.ones((width,height,3),dtype=np.uint8)*255
img_gray_1=np.zeros((width,height,3),dtype=np.uint8)

img_gray_1[:,:,:]=128

img_gray_2=np.random.randint(128,129,(width,height,3),dtype=np.uint8)
```

```python
fig=plt.figure(figsize=(10,7),facecolor='pink');
rows=1
columns=4

fig.add_subplot(rows,columns,1)
plt.imshow(img_black)
plt.axis('off')
plt.title("Black IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(img_white)
plt.axis('off')
plt.title("white IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(img_gray_1)
plt.axis('off')
plt.title("Gray1 IMAGE")

fig.add_subplot(rows,columns,4)
plt.imshow(img_gray_2)
plt.axis('off')
plt.title("Gray2 IMAGE")
```

```python
width,height,channel =300,420,3

img_flag_russian=np.zeros((width,height,channel),dtype=np.uint8)

img_flag_russian[:100,:]=[255,255,255]
img_flag_russian[100:200,:]=[0,0,160]
img_flag_russian[200:300,:]=[255,0,0]

plt.gcf().set_facecolor("gray")
mask_red =(img_flag_russian[:,:,0] == 255) & (img_flag_russian[:,:,1]==0)&(img_flag_russian[:,:,2]==0)

img_flag_russian[mask_red] =[255,255,0]
plt.imshow(img_flag_russian)
plt.axis('off')
plt.show()
```

```python
height, width, channel = 300, 420, 3


img_flag_belgium = np.zeros((height, width, channel), dtype=np.uint8)


img_flag_belgium[:, :140] = [0, 0, 0]        # Siyah (sol)
img_flag_belgium[:, 140:280] = [255, 255, 0] # Sar (orta)
img_flag_belgium[:, 280:] = [255, 0, 0]      # Krmz (sa)

mask_red = (img_flag_belgium[:, :, 0] == 0) & (img_flag_belgium[:, :, 1] == 0) & (img_flag_belgium[:, :, 2] == 0)


img_flag_belgium[mask_red] = [0, 0, 255]
plt.gcf().set_facecolor("gray")
plt.imshow(img_flag_belgium)
plt.axis('off')
plt.show()
```

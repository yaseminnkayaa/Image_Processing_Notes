```python
import matplotlib.pyplot as plt
import cv2
import numpy as npz

```

```python
my_original_img =cv2.imread("./binary.png",0)
my_kernel =np.ones((5,5),dtype=np.uint8)
my_erosion_img=cv2.erode(my_original_img,my_kernel,iterations=2)



fig = plt.figure(figsize=(10,7))
rows= 1
columns=2

fig.add_subplot(rows, columns,1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns,2)
plt.imshow(my_erosion_img, cmap="gray")
plt.axis('off')
plt.title("Eroıson ımage")
```

```python
my_dilation_img=cv2.dilate(my_original_img,my_kernel,iterations=2)



fig = plt.figure(figsize=(10,7))
rows= 1
columns=3

fig.add_subplot(rows, columns,1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns,2)
plt.imshow(my_erosion_img, cmap="gray")
plt.axis('off')
plt.title("Eroıson ımage")

fig.add_subplot(rows, columns,3)
plt.imshow(my_dilation_img, cmap="gray")
plt.axis('off')
plt.title("dilation  ımage")
```

```python
new_img =cv2.imread("./binary_with_text.jpeg",0)
white_noise=np.random.randint(0,2,size=new_img.shape[0:2])
white_noise =white_noise*255
```

```python
plt.imshow(white_noise, cmap="gray")
plt.axis('off')
plt.title("white and black  ımage")
```

```python
my_noised_img = new_img + white_noise


fig = plt.figure(figsize=(10,7))
rows= 1
columns=2



fig.add_subplot(rows, columns,1)
plt.imshow(new_img, cmap="gray")
plt.axis('off')
plt.title("Eroıson ımage")

fig.add_subplot(rows, columns,2)
plt.imshow(my_noised_img, cmap="gray")
plt.axis('off')
plt.title("dilation  ımage")
```

```python
my_opening_img =cv2.morphologyEx(my_noised_img.astype(np.float32),
                                 cv2.MORPH_OPEN,my_kernel)


fig = plt.figure(figsize=(10,7))
rows= 1
columns=3



fig.add_subplot(rows, columns,1)
plt.imshow(new_img, cmap="gray")
plt.axis('off')
plt.title("Eroıson ımage")

fig.add_subplot(rows, columns,2)
plt.imshow(my_noised_img, cmap="gray")
plt.axis('off')
plt.title("dilation  ımage")

fig.add_subplot(rows, columns,3)
plt.imshow(my_opening_img, cmap="gray")
plt.axis('off')
plt.title("dilation  ımage")
```

```python
my_black_noise =np.random.randint(0,2, size= my_original_img.shape[:2])
my_black_noise =my_black_noise* -255
plt.imshow(my_black_noise, cmap="gray")
plt.axis('off')
plt.title("my_black  ımage")
```

```python
my_black_noise_img=my_black_noise + new_img
```

```python
my_black_noise_img[my_black_noise_img <= -245]=0
```

```python
my_closing_img =cv2.morphologyEx(my_black_noise_img.astype(np.float32),
                                 cv2.MORPH_CLOSE,my_kernel)
```

```python

fig = plt.figure(figsize=(10,7))
rows= 1
columns=3



fig.add_subplot(rows, columns,1)
plt.imshow(new_img, cmap="gray")
plt.axis('off')
plt.title("Eroıson ımage")

fig.add_subplot(rows, columns,2)
plt.imshow(my_black_noise_img, cmap="gray")
plt.axis('off')
plt.title("vlack  ımage")

fig.add_subplot(rows, columns,3)
plt.imshow(my_closing_img, cmap="gray")
plt.axis('off')
plt.title("closed  ımage")
```

```python
my_gradient_img_1 =cv2.morphologyEx(my_original_img,
                                 cv2.MORPH_GRADIENT,my_kernel)

my_gradient_img_2 =cv2.morphologyEx(new_img,
                                 cv2.MORPH_GRADIENT,my_kernel)



fig = plt.figure(figsize=(10,7))
rows= 2
columns=2



fig.add_subplot(rows, columns,1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("original 1 ımage")

fig.add_subplot(rows, columns,2)
plt.imshow(my_gradient_img_1, cmap="gray")
plt.axis('off')
plt.title("my_gradient_img_1")

fig.add_subplot(rows, columns,3)
plt.imshow(new_img, cmap="gray")
plt.axis('off')
plt.title("original 2  ımage")


fig.add_subplot(rows, columns,4)
plt.imshow(my_gradient_img_2, cmap="gray")
plt.axis('off')
plt.title("my_gradient_img_2")



```
